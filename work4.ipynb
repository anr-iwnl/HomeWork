```python
n = 120
found = False
for k in range(1, int(n**(1/3)) + 1):
    if (k*(k+1)*(k+2) == n):
        found = True
        break
if found:
    print(k, k+1, k+2)
else:
    print("  ")
```

```python
import math

X, Y, Z, T = 3, 4, 5, 6

area_triangle = 0.5 * X * Y
perimeter = X + Y + Z + T
semi_perimeter = perimeter / 2

area_quadrilateral = math.sqrt(semi_perimeter * (semi_perimeter - X) * (semi_perimeter - Y) * (semi_perimeter - Z) * (semi_perimeter - T))

total_area = area_triangle + area_quadrilateral

print(total_area)
```
