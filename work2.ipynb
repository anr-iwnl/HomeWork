```python
def count_integers(m, n, p):
    integers = [x for x in [m, n, p] if isinstance(x, int)]
    return len(integers)

# Пример использования функции
m = 1.5
n = 2
p = 'abc'
result = count_integers(m, n, p)
print(f'Количество целых чисел: {result}')
```

```python
def check_sum_overflow(a, b):
    MAX_INT = 32767  # Максимальное целое число до переполнения

    # Проверяем, не превышает ли сумма максимальное значение
    if a + b > MAX_INT:
        print("Переполнение! Сумма больше 32767.")
    else:
        print(f"Сумма равна {a + b}.")

# Пример использования функции
A = 20000
B = 15000
check_sum_overflow(A, B)
```

```python
def analyze_two_digit_number(num):
    # Преобразуем число в строку для удобства работы с цифрами
    str_num = str(num)

    # Получаем первую и вторую цифры числа
    first_digit = int(str_num[0])
    second_digit = int(str_num[1])

    # Задача а): определить, какая цифра больше
    if first_digit > second_digit:
        print("Первая цифра больше второй.")
    elif first_digit < second_digit:
        print("Вторая цифра больше первой.")
    else:
        print("Цифры равны.")

    # Задача б): проверить, одинаковы ли цифры
    if first_digit == second_digit:
        print("Цифры одинаковы.")
    else:
        print("Цифры различны.")

# Пример использования функции
number = 23
analyze_two_digit_number(number)
```
