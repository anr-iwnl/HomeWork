```python
def calculate_expressions(x):
    # Вычисляем степени x
    x_squared = x ** 2  # x^2
    x_cubed = x_squared * x  # x^3

    # Выражение 1: -2x + 3x^2 - 4x^3
    expression1 = -2 * x + 3 * x_squared - 4 * x_cubed

    # Выражение 2: 1 + 2x + 3x^2 + 4x^3
    expression2 = 1 + 2 * x + 3 * x_squared + 4 * x_cubed

    return expression1, expression2

# Пример использования
x_value = 2
expression1_result, expression2_result = calculate_expressions(x_value)
print(f"Результат выражения 1: {expression1_result}")
print(f"Результат выражения 2: {expression2_result}")
```

```python
def both_are_odd(A, B):
    return A % 2 != 0 and B % 2 != 0

# Пример использования
A = 5
B = 9
if both_are_odd(A, B):
    print("Каждое из чисел A и B нечетное.")
else:
    print("Не каждое из чисел A и B нечетное.")
```
